```python
%matplotlib inline 
import matplotlib.pyplot as plt 
import numpy as np 

```

```python
np.random.seed(42)

x = 3*np.random.rand(200)
y = 20*x + 2*np.random.randn(200)

x = x.reshape(200,1)
y = y.reshape(200,1)
```

```python
print(x)
```

```python
print(y)
```

```python
xy = np.hstack([x,y])
print(xy.shape)
```

```python
plt.plot(xy[:,0], xy[:,1],'.')
plt.show()
```

```python
xy_centrado = xy - np.mean(xy, axis=0)
plt.plot(xy_centrado[:,0], xy_centrado[:,1],'.')
plt.show()
```

```python
autovalores , autovectores = np.linalg.eig(xy_centrado.T.dot(xy_centrado))
print(autovalores)
print('--')
print(autovectores)
```
